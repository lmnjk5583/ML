```python
# 옵션 설정: 모든 float 열을 소수점 2자리로 표시
pd.set_option('display.float_format', '{:.2f}'.format)

# 지수 표기법 억제하고 소수점 2자리로 고정
np.set_printoptions(suppress=True,    # 지수 표기법(e) 안 쓰기
                    precision=2,      # 소수점 아래 2자리
                    floatmode='fixed')  # 항상 소수점 2자리 표시 (0도 채움)
```

```python
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].values
```

```python
fish_input
```

```python
fish_target = fish[['Species']].values
```

```python
fish_target
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    fish_input, fish_target, random_state=42)
```

```python
train_input[:5]
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
train_scaled[:5]
```

```python
test_scaled[:5]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier(n_neighbors=3)
```

```python
kn.fit(train_scaled, train_target)
```

```python
# 학습데이터 성능
# 테스트데이터 성능
```

```python
print(kn.score(train_scaled, train_target))
print(kn.score(test_scaled, test_target))
```

```python
print(kn.classes_)
```

```python
print(kn.predict(test_scaled[:5]))
```

```python
kn.predict_proba(test_scaled[:5])
```
