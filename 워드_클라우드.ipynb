```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
import numpy as np
```

```python
with open('애국가.txt', 'r', encoding='utf-8') as f:
  text = f.read()
okt = Okt()
nouns = okt.nouns(text)
words = [n for n in nouns if len(n) > 1]
c = Counter(words)
```

```python
wc = WordCloud(font_path='C:/Users/mbc/AppData/Local/Microsoft/Windows/Fonts/NanumGothic.ttf',\
               width=400, height=400, scale=2.0, max_font_size=250)
gen = wc.generate_from_frequencies(c)
plt.figure()
plt.imshow(gen)

wc.to_file('애국가_워드클라우드.png')
```
