```python
from sklearn.model_selection import train_test_split
```

```python
56*0.75
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    perch_length, perch_weight, random_state=42)
```

```python
print(train_input.shape, test_input.shape)
```

```python
train_input = train_input.reshape(-1,1)
```

```python
test_input=test_input.reshape(-1,1)
```

```python
test_input
```

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knr = KNeighborsRegressor()
# k 최근접 이웃 회귀 모델의 훈련합니다.
knr.fit(train_input, train_target)
```

```python
# 종속 변수의 값이 독립 변수에 의해 얼마나 "결정(determined)"되거나 설명되는지를 나타앱니다.
# 즉, 무작위 변동이 아닌 모델로 예측 가능한 부분의 비율을 가리킵니다.
```

```python
# 회귀 성능 R^2(상관제곱) : 결정 계수
```

```python
knr.score(train_input, train_target)
```

```python
knr.score(test_input, test_target)
```

```python
# 훈련세트보다 테스트 세트의 점수가 높거나 두 점수가 모두 낮은 경우는 과속적합
```

```python
# 오차(실제값 - 예측값)
```

```python
# 실제값
test_target
```

```python
# 예측값
test_prediction = knr.predict(test_input)
```

```python
test_prediction
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test_target, test_prediction)
```
